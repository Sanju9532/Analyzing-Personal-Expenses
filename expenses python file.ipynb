```python
from faker import Faker
import pandas as pd
import random
from datetime import date
import calendar

fake = Faker(local='en_us')

def generate_monthly_expenses(month, year):
    num_transactions = 101
    data = []
    
    for i in range(num_transactions):
        transaction = {}
        start_date = date(year, month, 1)
        end_date = date(year, month, calendar.monthrange(year, month)[1])
        
        transaction['transaction_date'] = fake.date_between(start_date=start_date, end_date=end_date)
        transaction['Category'] = fake.random_element(elements=['Food', 'Transportation', 'Entertainment', 'School fees', 'Shopping', 'Groceries', 'Health', 'Travel', 'House Rent', 'Water bill', 'Electricity bill', 'wifi', 'Gas', 'Netflix', 'Audible', 'Disney Hotstar', 'YouTube', 'Spotify', 'Investment'])
        


        
        if transaction['Category'] == 'School fees':
            transaction['Amount Paid'] = 5000
        elif transaction['Category'] == 'Food':
            transaction['Amount Paid'] = 200
        elif transaction['Category'] == 'Transportation':
            transaction['Amount Paid'] = 1500
        elif transaction['Category'] == 'Entertainment':
            transaction['Amount Paid'] = 250
        elif transaction['Category'] == 'Shopping':
            transaction['Amount Paid'] = 1000
        elif transaction['Category'] == 'Groceries':
            transaction['Amount Paid'] = 500
        elif transaction['Category'] == 'Health':
            transaction['Amount Paid'] = 2000
        elif transaction['Category'] == 'Travel':
            transaction['Amount Paid'] = 400
        elif transaction['Category'] == 'House Rent':
            transaction['Amount Paid'] = 6000
        elif transaction['Category'] == 'Water bill':
            transaction['Amount Paid'] = 100
        elif transaction['Category'] == 'Electricity bill':
            transaction['Amount Paid'] = 700
        elif transaction['Category'] == 'wifi':
            transaction['Amount Paid'] = 600
        elif transaction['Category'] == 'Gas':
            transaction['Amount Paid'] = 1000
        elif transaction['Category'] == 'Netflix':
            transaction['Amount Paid'] = 300
        elif transaction['Category'] == 'Audible':
            transaction['Amount Paid'] = 299
        elif transaction['Category'] == 'Disney Hotstar':
            transaction['Amount Paid'] = 499
        elif transaction['Category'] == 'Spotify':
            transaction['Amount Paid'] = 350
        elif transaction['Category'] == 'YouTube':
            transaction['Amount Paid'] = 300
        elif transaction['Category'] == 'Investment':
            transaction['Amount Paid'] = 3000
        else:  
            transaction['Amount Paid'] = round(random.uniform(10, 1000), 2)
        
        transaction['Payment Mode'] = fake.random_element(elements=['Cash','Online', 'Debit card',  'Credit card'])
        transaction['Description'] = fake.sentence(nb_words=5)  

        
        if transaction['Category'] == 'School fees':
            transaction['Description'] = "halfyearly exam fees paid"
        elif transaction['Category'] == 'Groceries':
            transaction['Description'] = "1 week groceries are buyed"
        elif transaction['Category'] == "House Rent":
            transaction['Description'] = "Monthly payment is paid"
        elif transaction['Category'] == "Electricity bill":
            transaction['Description'] = "Monthly bill for electrical energy usage in home"
        elif transaction['Category'] == "Gas":
            transaction['Description'] = "Expenses for fuel used in heating and cooking"
        elif transaction['Category'] == "Shopping":
            transaction['Description'] = "Expenditures on clothing and personal items for everyday use"
        elif transaction['Category'] == "Health":
            transaction['Description'] = "Costs for medical care,health insurance and wellness services"
        elif transaction['Category'] == "Water bill":
            transaction['Description'] = "Charges for consumption of water services in a household"
        elif transaction['Category'] == "wifi":
            transaction['Description'] = "Costs associated with internet connectivity"
        elif transaction['Category'] == "Netflix":
            transaction['Description'] = "Subscription fees for movies and TV shows on the platform"
        elif transaction['Category'] == "Spotify":
            transaction['Description'] = "Subscription fees for music and podcasts from the service"
        else:
            transaction['Description'] = fake.sentence(nb_words=5) 



        
        
        
        transaction['Cashback'] = 0.0
        if transaction['Payment Mode'] == 'Online' and random.random() < 0.5:
            transaction['Cashback'] = round(transaction['Amount Paid'] * random.uniform(0.05, 0.20), 2)

        data.append(transaction)

    return pd.DataFrame(data)

# Example usage
year = 2023
for month in range(1, 13):
    df = generate_monthly_expenses(month, year)
    #print(f"Expenses for month {month}:\n", df)
```
